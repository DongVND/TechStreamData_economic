#CAC - Customer Acquisition Cost

##Calculate last month CRM expences

```python
monthly_expences_sheet_id = '10OGbaywwMIqKgnPGy8VDvpBVtjyqln47iYa2lFhI9Mw'
monthly_expences = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{monthly_expences_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
monthly_expences.head()
```

```python
last_month_expences = monthly_expences[monthly_expences['month'] == monthly_expences['month'].max()]
last_month_expences.head()
```

```python
crm_expence = last_month_expences[last_month_expences['item']== 'Salesforce']['amount'].values[0]
crm_expence
```

## Calculate last month Salaries

```python
payroll_sheet_id = '1c_WihqTZCQvNgxzmd-OwhR9i5diwtfxXVLyMn8R-Lp4'
payroll_monthly = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{payroll_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
payroll_monthly.head()
```

```python
last_month_payroll = payroll_monthly[payroll_monthly['month'] == payroll_monthly['month'].max()]
last_month_payroll.head()
```

We should calculate the payroll for marketing and sales since they are involve in the sell process. They include:
- Sales Associate
- Content Specialist

```python
last_month_payroll_mkt_sale_expence = last_month_payroll[last_month_payroll['department'].isin(['Sales','Marketing'])]['paid'].sum()
last_month_payroll_mkt_sale_expence
```

##Calculate last month Marketing costs

```python
marketing_spending_sheet_id = '1AZOIThOV4P-0eYDge53ZwumVkfkHoYPWxst3k3Bv87c'
marketing_spendings = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{marketing_spending_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
marketing_spendings.head()
```

```python
last_month_marketing_expences = marketing_spendings[marketing_spendings['date'].dt.month == marketing_spendings['date'].max().month]['spending'].sum()
last_month_marketing_expences
```

##Calculate last month Total Expences

```python
TotalExpences = crm_expence + last_month_payroll_mkt_sale_expence + last_month_marketing_expences
TotalExpences
```

##Calculate Number of Customers

```python
receipts_history_sheet_id = '1qayqML1zCKdmtzutkcy9LWvE6xFRm6TGBEVkHHJKIuE'
receipts_history = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{receipts_history_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
receipts_history.head()
```

```python
last_month_new_customer = receipts_history[(receipts_history['date'].dt.month == receipts_history['date'].max().month)&(receipts_history['new_customer'] == 1)]
last_month_new_customer.head()
```

```python
number_customer = len(last_month_new_customer)
number_customer
```

##Final : Calculate the CAC

```python
CAC = TotalExpences / number_customer
CAC
```

the CAC is about $1213.97

#Average Revenue Per User (ARPU)

```python
#Calculate total revenue in the last month
total_revenue_last_month = receipts_history[receipts_history['date'].dt.month == receipts_history['date'].max().month]['receipt_amount'].sum()
total_revenue_last_month
```

```python
#Total customer in the last month
number_customer_last_month = receipts_history[receipts_history['date'].dt.month == receipts_history['date'].max().month]['customer_id'].nunique()
number_customer_last_month
```

```python
#ARPU
ARPU = total_revenue_last_month / number_customer_last_month
ARPU
```

###Final: the ARPU is about 284

#COGS Cost of Goods Sold

Calculate everything spent on product development.(include: server cost and Atlassian Jira for technical support)

Part of Zoom and Slack cost should also be include. Since they were used by both dev team and back office
For simplicity, we assume 60% usage for COGS (dev team) and 40% usage for back office.



Production cost for server and software

```python
production_items = ['AWS Hosting', 'Google Cloud Storage','Atlassian Jira']
last_month_production_expences = last_month_expences[last_month_expences['item'].isin(production_items)]['amount'].sum()
```

```python
shared_software_items = ['Zoom','Slack']
last_month_shared_software_expences = last_month_expences[last_month_expences['item'].isin(shared_software_items)]['amount'].sum()
```

```python
server_n_software_expences = last_month_production_expences + last_month_shared_software_expences * 0.6
server_n_software_expences
```

```python
last_month_production_salary = last_month_payroll[last_month_payroll['department']=='Engineering']['paid'].sum()
last_month_production_salary
```

```python
COGS = server_n_software_expences + last_month_production_salary
COGS
```

#GROSS MARGIN

## Gross Margin calculation

```python
GrossMargin = (total_revenue_last_month - COGS) / total_revenue_last_month
GrossMargin
```

The result means Gross Margin is about 75.60%

##Gross Margin interpretation
- A 75% gross margin, or a 75% gross profit ratio, means that for every dollar of revenue, 75 cents is left over after accounting for the cost of producing the product or service.
- This remaining 75 cents can be used to cover operating expenses (like rent, salaries, marketing), and any remaining amount contributes to the company's net profit.
- For SaaS (Software as a Service) companies, a gross margin above 75% is often considered very good, indicating a healthy business model and efficient operations.
- A high gross margin also provides a business with more flexibility in terms of reinvesting in growth and managing unexpected costs.

In essence, a 75% gross margin suggests a company is effectively managing its production costs and has a strong foundation for profitability and growth.

#LTV Customer Lifetime Value

## Customer Lifespan calculation

```python
lifespan_sheet_id = '1by8tPHwOnq3uKYK2E7sA9VBUYoPM4p1Rnrm_Ss9cyHI'
lifespan_data = pd.read_excel(f'https://docs.google.com/spreadsheets/d/{lifespan_sheet_id}/export?format=xlsx', sheet_name='Sheet1')
lifespan_data.head()
```

```python
lifespan_data['lifespan_day'] = (lifespan_data['churn_date'] - lifespan_data['start_date']).dt.days
avg_lifespan_days = lifespan_data['lifespan_day'].mean()
avg_lifespan_month = (avg_lifespan_days / 30)
avg_lifespan_month
```

The customer average lifespan is around 9.84 months

##Final: LTV

```python
LTV = ARPU * avg_lifespan_month * GrossMargin
LTV
```

#LTV / CAC

```python
LTV/CAC
```

The "LTV/CAC" ratio is about 1.74

##**Solutions & Recommendations:**

The primary focus should be on improving the LTV/CAC ratio. This can be achieved by:

**1. Reduce Customer Acquisition Cost (CAC):**

  **Optimize Marketing Channels:**
  - Analyze ROI per Channel: Identify which specific marketing channels and campaigns are yielding the best results (most new customers for the lowest cost) and reallocate budget away from underperforming ones.
  - A/B Test Ad Creatives and Messaging: Continuously test different ads, landing pages, and calls to action to improve conversion rates and lower cost per lead/acquisition.
  - Refine Targeting: Ensure marketing efforts are precisely targeted at the ideal customer profile to reduce wasted spend.
  - Explore Organic Growth: Invest more in SEO, content marketing, and community building to acquire customers at a lower or zero direct acquisition cost.

  **Improve Sales Efficiency:**
  - Sales Funnel Optimization: Identify bottlenecks in the sales process that lead to lost leads or extended sales cycles. Streamline processes.
  - Sales Training: Enhance sales team skills to improve conversion rates from qualified leads.
  - Leverage CRM Better: Ensure the CRM is being used effectively to manage leads, track interactions, and improve follow-up.

**2. Increase Lifetime Value (LTV):**

  - Improve Customer Retention (Lower Churn): This is often the most impactful area for LTV improvement.
  - Onboarding Optimization: Ensure new customers are successfully onboarded and quickly realize value from the product to reduce early churn.
  - Proactive Customer Success: Implement a robust customer success program to engage with users, provide support, address issues, and ensure they are getting the most out of the product.
  - Gather Feedback: Actively solicit customer feedback (surveys, interviews) to understand pain points and areas for product improvement that would increase satisfaction and retention.
  - Loyalty Programs/Incentives: Consider programs that reward long-term customers.
  - Increase Average Revenue Per User (ARPU):
Upselling/Cross-selling: Introduce higher-tier plans with more features, or complementary products/services, to existing customers.
  - Pricing Optimization: Review pricing strategy to ensure it aligns with the value provided and market expectations. Could there be room for a slight price increase for new customers, or value-based pricing?
  - Feature Adoption: Encourage customers to use more features of the product to deepen their reliance and perceived value.

## **Immediate Action Plan:**
